Проверим, есть ли пустые значения у набора данных. Для этого, сначала создадим новый df, в котором будет проверка на то, является ли информация в ячейке пустой или нет

```python
is_null = anime_df.isnull()
```

Затем создадим фильтр, в которой будет значение true, если хотя бы в одной строке не будет данных

```python
row_with_null = is_null.any(axis=1)
```

Создаем новый df, используя фильтр

```python
rows_with_null = anime_df[row_with_null]
```

Выводим все строки с пустыми значениями некоторых полей

```python
rows_with_null
```

Сделаем фильтрацию по значениям полей

```python
# Фильтрация по рейтингу больше 9 и меньше 9.20
rating_df = anime_df[(anime_df["rating"] > 9) & (anime_df["rating"] < 9.20)]
```

```python
rating_df
```

```python
# Фильтрация по id = 9253 или 5114
episodes_df = anime_df[(anime_df["anime_id"] == 9253) | (anime_df["anime_id"] == 5114)]
```

```python
episodes_df
```

```python
# Фильтрация строк с помощью лямбда-функции (вывод названий с рейтингом больше 9.10 и меньше 9.20)
lambda_df = anime_df.loc[lambda anime_df: (anime_df["rating"] > 9.10) & (anime_df["rating"] < 9.20), "name"]
```

```python
lambda_df
```

Проведем сортировку отфильтрованного по рейтингу набора данных (по возрастанию name)

Выполним группировку (среднее значения рейтинга для каждого типа)

Подключаем еще один набор данных с рейтингом от пользователей

```python
rating_df = pd.read_csv("rating.csv")
rating_df.columns
```

Объединяем наборы данных по ключу anime_id

```python
Способ inner
```

```python
merge_out = anime_df.merge(rating_df, how = "inner", left_on = "anime_id", right_on = "anime_id")
```

```python
merge_out
```

Способ left

```python
merge_out = anime_df.merge(rating_df, how = "left", left_on = "anime_id", right_on = "anime_id")
merge_out
```

```python
Способ right
```

```python
merge_out = anime_df.merge(rating_df, how = "right", left_on = "anime_id", right_on = "anime_id")
merge_out
```

```python
Способ outer
```

```python
merge_out = anime_df.merge(rating_df, how = "outer", left_on = "anime_id", right_on = "anime_id")
merge_out
```
